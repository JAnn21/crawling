```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mydata= pd.read_csv('./csv/stastic.csv', encoding='CP949')

print(mydata.head())

Y=mydata['starbucks']
X=mydata.drop(['starbucks'], axis=1, inplace=False)
print(Y)
print(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=128)

lr = LinearRegression()
print(lr.fit(X_train, Y_train))

Y_predict = lr.predict(X_test)
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
coef = pd.Series(data=np.round(lr.coef_, 1), index=X.columns)
print(coef.sort_values(ascending=False))

fig, axs = plt.subplots(figsize=(16, 16), ncols=3, nrows=3)
x_features = ['caffe_pascucci','caffebene','tomandtoms','hollys','twosome','yogerpresso','ediya','income','pop']
print("d")
for i, feature in enumerate(x_features):
    row = i//3
    col = i%3
    #print("i = %d , feature = %s ,  row = %d, col = %d" %(i, feature, row, col))
    sns.regplot(x=feature, y='starbucks', data=mydata, ax=axs[row][col])
```
